```python
import numpy as np
import matplotlib.pyplot as plt

# Provided vectors
a_geom= np.array([-5.08642212e+02, -6.90024287e+00,  1.60410743e-01])
a_grav= np.array([-7.34957928e-03, -6.18409635e-03,  1.04952918e-06])

fig = plt.figure(figsize=(7, 7))
ax = fig.add_subplot(111, projection='3d')

# Origin for both vectors
origin = np.zeros(3)

# Plot a_geom
ax.quiver(*origin, *a_geom, color='r', length=1.0, normalize=True, label='a_geom')

# Plot a_grav
ax.quiver(*origin, *a_grav, color='b', length=1.0, normalize=True, label='a_grav')

# --- How do I zoom in? ---
# If you want to zoom in to see both vectors better, 
# set the axis limits manually to a smaller range.
# For example, try a range around zero that fits both vectors:
# Since a_grav is much smaller than a_geom, let's zoom-in around the origin (e.g., [-0.03, 0.03]):

zoom_lim = 0.5  # Try 0.01, 0.03, or something suitable for a_grav
ax.set_xlim([-zoom_lim, zoom_lim])
ax.set_ylim([-zoom_lim, zoom_lim])
ax.set_zlim([-zoom_lim, zoom_lim])

# Optionally, plot origin for reference
ax.scatter(0,0,0, color='k', s=20)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Zoomed: a_geom (red) and a_grav (blue) vectors\n(axes set to [-0.03, 0.03])')
ax.legend()
plt.show()
```
